<b>value_counts():</b> it returns a series object, counting all the unique values. <br>
The first value is the most frequently occurring element, secong the second most occurring element and so on. <br>
The order of the series can be reversed by setting the ascending flag to true. <br>
The dropna parameter is initially set to True so it will not return the NA values. <br>


```python
#import the pandas library
import pandas as pd
```

```python
#read the csv file
olympicsDataFrame = pd.read_csv('data/olympics.csv')
```

```python
olympicsDataFrame.head()
```

```python
#let's see in which year the most medals were presented
olympicsDataFrame.Edition.value_counts()
```

```python
#now let's check the number of medals presented to men and the number of medals presented to women throughout
#the history of the olympics
#Additionaly we'll use the ascending flag
#Remember that by default the na values aren't counted
olympicsDataFrame.Gender.value_counts(ascending=True)
```

<b>sort_values():</b> sorts the values in a series. <br>
By default, values are sorted in ascending order. <br>
Bt default, the missing data or NaN are put at the end. <br>


```python
#Let's sort by the athelete's names
olympicsDataFrame['Athlete'].sort_values()
```

```python
#Now let's sort based on Edition and Athele name
olympicsDataFrame.sort_values(by=['Edition','Athlete'])
```

<b>Boolean Indexing:</b> Boolean vectors or conditionals can be used to filters data. <br>
Based on a condition , pass series of true and false values to a series or data frame and display the rows has true values.<br>
We use the symbols (& for and) (| for or) and (! for not)
    

```python
#Let's display the atheletes who have won gold medals
olympicsDataFrame[olympicsDataFrame['Medal']=='Gold']
```

```python
#Now let's find all the women athletes who have won gold medals
olympicsDataFrame[(olympicsDataFrame['Medal'] =='Gold') & (olympicsDataFrame['Gender'] == 'Women')]
```

<b>String handling:</b> available to every series using str attribute. <br>
Using series.str you have access to a lot of string methods.

```python
#let us find the records of athlete with the first name Florence
olympicsDataFrame[olympicsDataFrame['Athlete'].str.contains('Florence')]
```

```python
#Let us find the events in which Jessie Owens won a medal
jesse_owens_events = olympicsDataFrame[olympicsDataFrame['Athlete']=='OWENS, Jesse']
jesse_owens_events
```
