```python
import pandas as pd
import matplotlib.pyplot as plt
#merger command
%matplotlib inline
import seaborn as sns
```

The plot command by default displays a line graph. <br>
You can change the type of graph by using the kind= parameter. <br> 

```python
olympicsDataFrame = pd.read_csv('data/olympics.csv')
olympicsDataFrame.head()
```

```python
#What were the different sports in the first olympics? Plot them using different graphs.
first_olympics_data = olympicsDataFrame[olympicsDataFrame['Edition'] == 1896]
first_olympics_data
```

```python
first_olympics_sports = first_olympics_data['Sport'].value_counts()
first_olympics_sports
```

```python
#line graph
first_olympics_sports.plot()
```

```python
#Bar graph
first_olympics_sports.plot(kind = 'bar')
```

```python
#Horizontal bar plot
first_olympics_sports.plot(kind = 'barh')
```

```python
#Pie Chart
first_olympics_sports.plot(kind = 'pie')
```

```python
#You can specify the color of the chart by using the color attribute
#https://matplotlib.org/stable/gallery/color/named_colors.html
first_olympics_sports.plot(kind = 'bar', color = 'plum')
```

```python
#figsize is a tuple attribute which specifies the width and height in inches
first_olympics_sports.plot(figsize=(10,3))
```

Colormaps help to better represent your data.<br>
There are three classes of colormaps:<br>
1. Sequential: For info that is ordered<br>
2. Diverging: when info deviates around the middle value<br>
3. Qualitative: when info does not have any ordering or relationship<br>


```python
first_olympics_sports.plot(kind = 'pie', colormap = 'Pastel1')
```

<b>Seaborn</b> is a vizualisation library based on matplotlib. <br>
1. It provides beautiful statistical plots<br>
2. It is a complement and not a substitute for matplotlib <br>
3. Integrates well with pandas <br>


```python
#Let's find how many medals have been won by men and women in the history of the olympics. How many gold, silver,
#and bronze were won by each gender?
sns.countplot(x='Medal', data= olympicsDataFrame, hue='Gender')
```

```python
#Let's plot the number of medals achieved by the Chinese team(men and women) in Beijing 2008 using Matplotlib
china_2008_data = olympicsDataFrame[(olympicsDataFrame['NOC'] == 'CHN') & (olympicsDataFrame['Edition'] == 2008)]
china_2008_data
```

```python
china_2008_gender = china_2008_data['Gender'].value_counts()
china_2008_gender
```

```python
china_2008_gender.plot(kind = 'bar')
```

```python
sns.countplot(x = 'Gender', data = china_2008_data)
```

```python
sns.countplot(x = 'Gender', data = china_2008_data, palette = 'bwr' )
```

```python
#Let's now plot the number of medals for each gender
sns.countplot(x = 'Medal', data = china_2008_data, hue = 'Gender')
```

```python
sns.countplot(x = 'Medal', data = china_2008_data, hue = 'Gender', palette = 'bwr', order = ['Gold','Silver','Bronze'])
```
