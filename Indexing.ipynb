```python
import pandas as pd
```

The Index object is an immutable array. <br>
Indexing allows you to access a row or column using a label. <br>
Indexes are immutable. <br>

```python
olympicsDataFrame = pd.read_csv('data/olympics.csv')
```

```python
#An index has its own type
type(olympicsDataFrame.index)
```

The set_index() allows us to determine which of the series is going to be the index. <br>


```python
#Let us set the 'Athelte' as the index
#Method1:
olympicsDataFrame.set_index('Athlete')
```

```python
#We'd expect the index column to have changed. But this is not the case.
olympicsDataFrame.head()
```

```python
#In order to reflect changes in original DataFrame, we need to set the inplace attribute to True
olympicsDataFrame.set_index('Athlete', inplace=True)
```

```python
olympicsDataFrame.head()
```

```python
#Reset index to its default integer based index
olympicsDataFrame.reset_index(inplace = True)
```

```python
olympicsDataFrame
```

```python
#Method 2 of changing index:
athlete_indexed_DataFrame = olympicsDataFrame.set_index('Athlete')
athlete_indexed_DataFrame.head()
```

sort_index allows for all the items to be sorted based on its index

```python
#Lets sort the items based on the athletes name
athlete_indexed_DataFrame.sort_index(inplace=True)
athlete_indexed_DataFrame.head()
```

```python
#Lets sort the items based on the athletes name in descending order
athlete_indexed_DataFrame.sort_index(inplace = True, ascending = False)
athlete_indexed_DataFrame
```

loc[] is a label based indexer for selection by label. <br>
loc[] will raise a KeyError when the items are not found. <br>


```python
#Let us find the records of Usain Bolt
athlete_indexed_DataFrame.loc['BOLT, Usain']
```

```python
#Another way to do this
olympicsDataFrame[olympicsDataFrame['Athlete'] == 'BOLT, Usain']
```

```python
#Yet another way of doing this
olympicsDataFrame.loc[olympicsDataFrame['Athlete'] == 'BOLT, Usain']
```

iloc[] is used for selection by integer index

```python
#Suppose we want the 1700th record
olympicsDataFrame.iloc[1700]
```

```python
#To fetch multiple records
olympicsDataFrame.iloc[[1200, 4562, 12003, 24]]
```

```python
#Suppose we want records 1 to 10
olympicsDataFrame.iloc[1:11]
```

```python
#Lets plot the total number of medals awarded at each of the Olympic games throughout history
medals_yearly_count_data = olympicsDataFrame['Edition'].value_counts().sort_index()
medals_yearly_count_data
```

```python
medals_yearly_count_data.plot(kind = 'bar', figsize = (10, 10), )
```

```python
#Lets find which countries did not win a medal in 2008 and how many countries were there.
participatingCountriesDataFrame = pd.read_csv(r'data/Summer Olympic medallists 1896 to 2008 - IOC COUNTRY CODES.csv')
participatingCountriesDataFrame
```

```python
#To check the difference between content in 'Country' and 'Country.1'
participatingCountriesDataFrame[participatingCountriesDataFrame['Country'] != participatingCountriesDataFrame['Country.1']]
```

```python
#Let us set 'Int Olympic Committee code' as the index of participatingCountriesDataFrame
participatingCountriesDataFrame.set_index('Int Olympic Committee code', inplace = True)
participatingCountriesDataFrame
```

```python
olympics2008_data = olympicsDataFrame[olympicsDataFrame['Edition'] == 2008]
olympics2008_data['NOC'].value_counts()
```

```python
countries_medal_2008 = olympics2008_data['NOC'].value_counts()
countries_medal_2008
```

```python
participatingCountriesDataFrame['medal_2008'] = countries_medal_2008
participatingCountriesDataFrame
```

```python
countries2008_no_medal = participatingCountriesDataFrame[(participatingCountriesDataFrame['medal_2008'].isnull())]
```
